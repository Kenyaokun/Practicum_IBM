### After exploration, I tend to merge 'luad_mskcc_2020_clinical_data(604)',  'luad_tcga_clinical_data(586)' and 'nsclc_tcga_broad_2016_clinical_data(1145)' 

```python
luad_mskcc = luad_mskcc_2020_clinical_data[['Study ID',
 'Patient ID',
 'Sample ID',
 'Age At Surgery',
 'Cancer Type',
 'Cancer Type Detailed',
 'Fraction Genome Altered',
 'Mutation Count',
 'Oncotree Code',
 'Overall Survival (Months)',
 'Overall Survival Status',
 'Number of Samples Per Patient',
 'Sex',
 'Person Cigarette Smoking History Pack Year Value',
 'TMB (nonsynonymous)']]
```

```python
luad_mskcc_new = luad_mskcc.rename(columns = {'Age At Surgery' : 'Age'})

luad_mskcc_new
```

```python
nsclc_tcga_broad = nsclc_tcga_broad_2016_clinical_data[['Study ID',
 'Patient ID',
 'Sample ID',
 'Diagnosis Age',
 'Cancer Type',
 'Cancer Type Detailed',
 'Fraction Genome Altered',
 'Mutation Count',
 'Oncotree Code',
 'Overall Survival (Months)',
 'Overall Survival Status',
 'Number of Samples Per Patient',
 'Sex',
 'Person Cigarette Smoking History Pack Year Value',
 'TMB (nonsynonymous)']]
```

```python
nsclc_tcga_broad_new = nsclc_tcga_broad.rename(columns = {'Diagnosis Age' : 'Age'})
nsclc_tcga_broad_new
```

```python
nsclc_tcga_broad_2016 = nsclc_tcga_broad_2016_clinical_data[[
 'Study ID',
 'Patient ID',
 'Sample ID',
 'Diagnosis Age',
 'Cancer Type',
 'Cancer Type Detailed',
 'Fraction Genome Altered',
 'Mutation Count',
 'Oncotree Code',
 'Overall Survival (Months)',
 'Overall Survival Status',
 'Number of Samples Per Patient',
 'Sex',
 'Person Cigarette Smoking History Pack Year Value',
 'TMB (nonsynonymous)'
]]
```

```python
nsclc_tcga_broad_2016_new = nsclc_tcga_broad_2016.rename(columns = {'Diagnosis Age' : 'Age'})
nsclc_tcga_broad_2016_new
```

```python
lung_cancer = pd.concat([luad_mskcc_new, nsclc_tcga_broad_new, nsclc_tcga_broad_2016_new], axis=0)

lung_cancer.reset_index(drop = True)



```
